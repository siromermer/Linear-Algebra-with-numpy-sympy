## Vector addition & subtraction

* Vector addition involves adding corresponding components of two or more vectors to create a new vector.
* Vector subtraction is similar to addition but involves subtracting corresponding components.

<br>

#### 1.1.4 : Draw v=[4,1] and w=[-2,2] and v+w and v-w in a single xy plane. 

```python
# Define the vectors
vector_1 = np.array([4, 1])
vector_2 = np.array([-2, 2])

vector_sum=np.add(vector_1,vector_2)
vector_subtract=np.subtract(vector_1,vector_2)

# Create a new figure for the 2D plot
fig, ax = plt.subplots()

# Plot the vector
ax.quiver(0, 0, vector_1[0], vector_1[1], angles='xy', scale_units='xy', scale=1, label='vector 1 [4,1]',color="b")
ax.quiver(0, 0, vector_2[0], vector_2[1], angles='xy', scale_units='xy', scale=1, label='vector 2 [-2,2]',color="black")
ax.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, label='vector1 + vector2',color="y")
ax.quiver(0, 0, vector_subtract[0], vector_subtract[1], angles='xy', scale_units='xy', scale=1, label='vector1 - vector2',color="r")


# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Set the limits of the plot to show all four quadrants
ax.set_xlim([-7, 7])
ax.set_ylim([-7, 7])

# Add a legend
ax.legend()

# Set equal aspect ratio
ax.set_aspect('equal')

# Draw grid
ax.grid(True)

plt.title("Addition & Subtraction")

# Show the plot
plt.show()
```

#### Additional Example --> Vector addition / subtraction in 3 dimension

```python
# Define the vectors
vector_1 = np.array([1, 0 ,0])
vector_2 = np.array([0, 2, 3])

# Create a new figure for the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the origin
origin = np.array([0, 0, 0])

# vector 1 and vector 2
ax.quiver(*origin, *vector_1, color='b', label='Vector [1, 0 ,0]')
ax.quiver(*origin, *vector_2, color='r', label='Vector [0, 2, 3]')

# vector1  + vector2
vector_3 = np.add(vector_1, vector_2)
ax.quiver(*origin, *vector_3, color='y', label='Vector1 + Vector2 ')

# vector1  - vector2
vector_4 = np.subtract(vector_1, vector_2)
ax.quiver(*origin, *vector_4, color='y', label='Vector1 - Vector2 ')

# Set labels for the axes

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the limits of the plot
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

## Dot Product

#### Dot Product :  It also known as the scalar product, is an operation that takes two vectors and returns a scalar.
Applications :
* Finding Angle Between Vectors
* Finding if vectors are Orthogonal
* For projecting one vector onto another.

<br>

#### 1.2.1 : Calculate the dot products u.v and u.w and u.(v+w) and w.v

```python
u = np.array([-6,8])
v = np.array([4,3])
w = np.array([1,2])

print(np.dot(u, v)) #  u * v
print(np.dot(u, w)) # u * w
print(np.dot(u, np.add(v,w))) # u * (v + w) 
print(np.dot(w, v)) #  w * v
```

## Unit Vector

#### A unit vector is a vector with a magnitude (length) of 1. It is used to indicate direction without considering the magnitude
* (unit vector) u = u / ||u||

 <br>

```python
u = np.array([-6,8])

total = np.square(u)[0] + np.square(u)[1]

unit_x = u[0] / np.sqrt(total)
unit_y = u[1] / np.sqrt(total)
 
print(unit_x,unit_y)
```

```python
# Create a new figure for the 2D plot
fig, ax = plt.subplots()

# Plot the vector
ax.quiver(0, 0, unit_x, unit_y, angles='xy', scale_units='xy', scale=1, label='Unit Vector',color="y",width=0.02)

ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, label='Vector [-6, 8]',color="b")


# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Set the limits of the plot to show all four quadrants
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# Add a legend
ax.legend()

# Set equal aspect ratio
ax.set_aspect('equal')

# Draw grid
ax.grid(True)

plt.title("Unit vector of vector u and vector u")

# Show the plot
plt.show()
```

## Find Angle between 2 Vector

#### Find the angle between two vectors using the dot product formula : uv = uv . cos 
* uv is the dot product of vectors
* u and v are the magnitudes (lengths)
*  is the angle between the vectors

<br>

#### 1.2.7.a : Find the angle  between these pairs of vectors : v = [1, 3**0.5 ] , w= [1,0] 

```python
# Find angle between v and w
v = np.array([1,np.sqrt(3)]) # square root of 3
w = np.array([1,0])

# cos0 = a.b / |a||b|

# Compute the dot product
dot_product = np.dot(v, w)

# Compute the magnitudes of the vectors
mag_v = np.linalg.norm(v)
mag_w = np.linalg.norm(w)

# Compute the cosine of the angle
cos_theta = dot_product / (mag_v * mag_w)

# Compute the angle in radians
angle_radians = np.arccos(cos_theta)

# Convert the angle to degrees (optional)
angle_degrees = np.degrees(angle_radians)

print(angle_degrees)
```

```python
# Create a new figure for the 2D plot
fig, ax = plt.subplots()

# Plot the vector
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='v',color="b")
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, label='w',color="y")


# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Set the limits of the plot to show all four quadrants
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])

# Add a legend
ax.legend()

# Set equal aspect ratio
ax.set_aspect('equal')

# Draw grid
ax.grid(True)

plt.title(f"angle between u and w : {round(angle_degrees,2)}")

# Show the plot
plt.show()
```

## Columns Space , Null Space , Pivot Columns , RREF form

<br>

##### In chapter 3 , this topics will explained in depth

* Column Space: The set of all possible linear combinations of the columns of a matrix, representing the span of its column vectors.

* Null Space: The set of all vectors that, when multiplied by the matrix, result in the zero vector, representing the solutions to the homogeneous equation --> Solves x for Ax=0

* Pivot Columns: The columns of a matrix corresponding to the leading entries (pivots) in its row echelon form, which are linearly independent and form a basis for the column space.



<br>

#### 1.3.2.a : Find a combination of the columns that produces (0,0,0)  . The trivial combination (zero times every column) is not allowed. Columns are --> a=[1,4,7] , b=[2,5,8] , c=[3,6,9]

```python
a = np.array([1,4,7])
b = np.array([2,5,8])
c = np.array([3,6,9])

A = np.column_stack((a, b, c))

A  = sympy.Matrix(A)
A
```

#### Find nullspace --> Ax=0  , x is nullspace

```python
null_space = A.nullspace() # 2b -a = c 
null_space
```

<br>

#### Reduce row echelond form , pivot columns

```python
rref, pivot_columns = sympy.Matrix(A).rref() # returns --> first one is reduced row echelon form, and the second is a tuple of indices of the pivot columns
rref  # as you can see from rref , column1 and column2 are pivot columns 
```

```python
pivot_columns # these are pivot columns
```
